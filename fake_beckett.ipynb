```python
# import io module for encoding help
import io
```

```python
with io.open("beckettpoems.txt", encoding='utf8', errors='ignore') as f:
    with io.open("beckett_samples_1.txt", encoding='utf8', errors='ignore') as e:
        poems = f.read()
        samples = e.read()
```

```python
# clean up beckett poems and turn into lists
poems = poems.split("\n")
poems = list(filter(None, poems))
```

```python
#clean up fake beckett samples and turn into lists
samples = samples.split("\n")
samples = list(filter(None, samples))
```

```python
# remove sample headers
def remove_sample(samp):
    return len(samp) < 80
samples = list(filter(remove_sample, samples))
```

```python
'''creating a list of unique lines that exist in the
GPT-2 generated beckett list that do not exist in the actual
beckett poem list'''
fake_beckett = list(set(samples) - set(poems))
```

```python
print("\n".join(fake_beckett))
```
